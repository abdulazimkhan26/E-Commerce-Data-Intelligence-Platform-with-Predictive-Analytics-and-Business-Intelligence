<h1>Loading into postgreSql</h1>

```python
import os
from os import path
import pandas as pd
```

```python
path = "./cleaned_ds"
os.makedirs(path, exist_ok=True)
```

```python
customers = pd.read_csv(os.path.join(path, "customer_clean.csv"))
order_items = pd.read_csv(os.path.join(path, "order_items_clean.csv"))
orders = pd.read_csv(os.path.join(path, "orders_clean.csv"))
payment = pd.read_csv(os.path.join(path, "payments_clean.csv"))
product = pd.read_csv(os.path.join(path, "products_clean.csv"))
review = pd.read_csv(os.path.join(path, "reviews_clean.csv"))
seller = pd.read_csv(os.path.join(path, "sellers_clean.csv"))
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "postgresql+psycopg2://postgres:260849@localhost:5433/Olist",
    pool_pre_ping=True
)

with engine.begin() as conn:
    customers.to_sql(
        "customer",
        conn,
        if_exists="append",
        index=False,
        method='multi',
        chunksize=1000
    )
```

```python
with engine.begin() as conn:
    orders.to_sql(
        "orders",
        conn,
        if_exists="append",
        index=False,
        method='multi',
        chunksize=1000
    )
```
